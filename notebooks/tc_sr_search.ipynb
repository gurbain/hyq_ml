# Training with Open Loop Oscillator

#### 1. Declare the network structure and the training parameters

```python
%load_ext autoreload
%autoreload
%matplotlib inline

import network
import utils

# Readout layer and backprop algo
nn_layers       = [[('relu', 40)],
                   [('relu', 40)],
                   [('relu', 40)]
                  ]
batch_size      = 2048
max_epochs      = 100
stop_patience   = 150
regularization  = 0.001
metric          = "mae"
optimizer       = "adam"

# ESN
n_res           = 150
n_read          = 150
damping         = 0.1
noise           = 0.01
in_mask         = None  # No output is injected in the ESN
out_mask        = None  # All readouts ouputs are fed back in the ESN

# Other
data_file       = "../data/sims/tc.pkl"
save_folder     = "../data/nn_learning/" + utils.timestamp()
verbose         = -1
utils.mkdir(save_folder)
```

#### Try different spectral radius values

```python
# 0.1
del sparsity, spectral_radius, nn
sparsity        = 0.
spectral_radius = 0.1
nn = network.NN(nn_layers=nn_layers, optim=optimizer, metric=metric, batch_size=batch_size, 
                max_epochs=max_epochs, regularization=regularization, esn_n_res=n_res,
                esn_n_read=n_read, esn_in_mask=in_mask, esn_out_mask=out_mask,
                esn_spec_rad=spectral_radius, esn_damping=damping, esn_sparsity=sparsity,
                esn_noise=noise, data_file=data_file, save_folder=save_folder, checkpoint=False,
                verbose=verbose,random_state=42)
loss, acc = nn.train(evaluate=True, plot_test_states=False, plot_test=True, win=1000)
```
